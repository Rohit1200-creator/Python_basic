```python
#2.Describe the role of predefined keywords in Python and provide examples of how they are used in a program.  


##Ans.In Python, predefined keywords are reserved words that have special meanings and are used to define the structure and logic of a program. They cannot be used as variable names or identifiers. Here are some examples of predefined keywords in Python and how they are used:

1. if, else, elif: Used for conditional statements to control the flow of a program.

Example:

x = 5
if x > 10:
    print("x is greater than 10")
else:
    print("x is less than or equal to 10")


1. for, while: Used for loops to iterate over a sequence or repeat a block of code.

Example:

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)


1. def: Used to define a function.

Example:

def greet(name):
    print("Hello, " + name + "!")
greet("John")


1. class: Used to define a class.

Example:

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def greet(self):
        print("Hello, my name is " + self.name + " and I am " + str(self.age) + " years old.")


1. try, except: Used for exception handling to catch and handle errors.

Example:

try:
    x = 5 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")


1. import: Used to import modules or libraries.

Example:

import math
print(math.pi)


1. return: Used to exit a function and return a value.

Example:

def add(x, y):
    return x + y
result = add(5, 3)
print(result)


1. global: Used to declare a variable as global.

Example:

x = 5
def change_x():
    global x
    x = 10
change_x()
print(x)


1. nonlocal: Used to declare a variable as nonlocal in a nested function.

Example:

def outer():
    x = 5
    def inner():
        nonlocal x
        x = 10
    inner()
    print(x)
outer()


These are just a few examples of the many predefined keywords in Python. Understanding how to use these keywords is essential to writing effective and efficient Python programs.
```

```python
#3.Compare and contrast mutable and immutable objects in Python with examples.  


##Ans.In Python, objects can be classified as either mutable or immutable.

Mutable Objects:

- Can be changed after creation
- Modifications affect the original object
- Examples: lists, dictionaries, sets, user-defined classes

Example:

my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]


Immutable Objects:

- Cannot be changed after creation
- Modifications create a new object
- Examples: integers, floats, strings, tuples

Example:

my_string = "hello"
my_string += " world"
print(my_string)  # Output: "hello world"
print(id(my_string))  # Different ID, new object created


Key differences:

- Modifiability: Mutable objects can be modified in-place, while immutable objects require creating a new object.
- Memory usage: Immutable objects can share memory, reducing memory usage.
- Thread safety: Immutable objects are thread-safe, as they cannot be modified concurrently.

When to use each:

- Mutable objects: When you need to modify data in-place, such as updating a list or dictionary.
- Immutable objects: When you need to ensure data integrity, thread safety, or memory efficiency, such as working with strings or integers.

Remember, understanding the difference between mutable and immutable objects is crucial for writing efficient, safe, and effective Python code.
```

```python
#4.Discuss the different types of operators in Python and provide examples of how they are used.  


##Ans.Python has several types of operators, including:

1. Arithmetic Operators:
    - + (addition)
    - - (subtraction)
    - * (multiplication)
    - / (division)
    - ** (exponentiation)
    - // (floor division)
    - % (modulus)

Example:

a = 5
b = 2
print(a + b)  # Output: 7
print(a ** b)  # Output: 25


1. Comparison Operators:
    - == (equal to)
    - != (not equal to)
    - > (greater than)
    - < (less than)
    - >= (greater than or equal to)
    - <= (less than or equal to)

Example:

a = 5
b = 2
print(a > b)  # Output: True
print(a == b)  # Output: False


1. Logical Operators:
    - and (logical and)
    - or (logical or)
    - not (logical not)

Example:

a = True
b = False
print(a and b)  # Output: False
print(a or b)  # Output: True
print(not a)  # Output: False


1. Assignment Operators:
    - = (assignment)
    - += (addition assignment)
    - -= (subtraction assignment)
    - *= (multiplication assignment)
    - /= (division assignment)
    - **= (exponentiation assignment)

Example:

a = 5
a += 2
print(a)  # Output: 7


1. Membership Operators:
    - in (membership)
    - not in (non-membership)

Example:

my_list = [1, 2, 3]
print(2 in my_list)  # Output: True
print(4 not in my_list)  # Output: True


1. Identity Operators:
    - is (identity)
    - is not (non-identity)

Example:

a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)  # Output: False (different objects)
print(a == b)  # Output: True (same value)


These operators help you perform various operations, comparisons, and assignments in Python, making your code more efficient and effective.
```

```python
#5.Explain the concept of type casting in Python with examples.


##Ans.Type casting in Python is the process of converting a variable's data type into another data type. This is useful when you need to use a variable in a context where a different data type is required.

Here are some examples of type casting in Python:

1. Int to Float:

x = 5  # integer
y = float(x)  # convert to float
print(y)  # Output: 5.0


1. Float to Int:

x = 5.5  # float
y = int(x)  # convert to integer
print(y)  # Output: 5


1. String to Int:

x = "123"  # string
y = int(x)  # convert to integer
print(y)  # Output: 123


1. String to Float:

x = "123.45"  # string
y = float(x)  # convert to float
print(y)  # Output: 123.45


1. Int to String:

x = 123  # integer
y = str(x)  # convert to string
print(y)  # Output: "123"


1. Float to String:

x = 123.45  # float
y = str(x)  # convert to string
print(y)  # Output: "123.45"


1. Boolean to Int:

x = True  # boolean
y = int(x)  # convert to integer
print(y)  # Output: 1


1. Boolean to String:

x = True  # boolean
y = str(x)  # convert to string
print(y)  # Output: "True"

```

```python
#6.How do conditional statements work in Python? Illustrate with examples.


##Ans.Conditional statements in Python are used to execute different blocks of code based on certain conditions. They are also known as decision-making statements. Here's how they work:

1. If Statement: Executes a block of code if a condition is true.

Example:

x = 5
if x > 10:
    print("x is greater than 10")
else:
    print("x is less than or equal to 10")


1. If-Else Statement: Executes different blocks of code based on multiple conditions.

Example:

x = 5
if x > 10:
    print("x is greater than 10")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")


1. If-Elif-Else Statement: Executes different blocks of code based on multiple conditions.

Example:

x = 5
if x > 10:
    print("x is greater than 10")
elif x == 5:
    print("x is equal to 5")
elif x < 0:
    print("x is less than 0")
else:
    print("x is between 0 and 10")


1. Nested If Statements: If statements inside if statements.

Example:

x = 5
if x > 10:
    if x > 20:
        print("x is greater than 20")
    else:
        print("x is between 10 and 20")
else:
    print("x is less than or equal to 10")


1. One-Line If Statements: If statements in a single line.

Example:

x = 5
print("x is greater than 10") if x > 10 else print("x is less than or equal to 10")


Remember, conditional statements are a fundamental part of programming, and mastering them is essential for writing effective and efficient code.
```

```python
#7.Describe the different types of loops in Python and their use cases with examples.  


##Ans.Python has two primary types of loops: for loops and while loops.

For Loops

- Used for iterating over a sequence (list, tuple, string, dictionary, set)
- Execute a block of code for each item in the sequence

Example:

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)


While Loops

- Used for executing a block of code as long as a condition is true
- Continue iterating until the condition becomes false

Example:

i = 0
while i < 5:
    print(i)
    i += 1


Nested Loops

- Loops inside loops
- Used for iterating over multiple sequences or complex data structures

Example:

colors = ["red", "green", "blue"]
shapes = ["circle", "square", "triangle"]
for color in colors:
    for shape in shapes:
        print(f"{color} {shape}")


Infinite Loops

- Loops that run indefinitely
- Used for tasks that need to run continuously, like monitoring or listening for events

Example:

while True:
    # do something
    pass


Break and Continue Statements

- break: Exit the loop prematurely
- continue: Skip the current iteration and move to the next one

Example:

for i in range(10):
    if i == 5:
        break
    print(i)


Loop Control Statements

- range(): Generate a sequence of numbers for looping
- enumerate(): Loop over both index and value of a sequence
- zip(): Loop over multiple sequences in parallel

Example:

names = ["John", "Jane", "Jim"]
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(f"{name} is {age} years old")


Understanding these loop types and their use cases is essential for writing efficient and effective Python code.
```
